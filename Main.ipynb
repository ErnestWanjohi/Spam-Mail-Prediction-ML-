```python
#Imprting dependencies
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset to a pandas DataFrame
raw_mail_data = pd.read_csv('D:\\Spam Mail Prediction (ML)\\mail_data.csv')
```

```python
raw_mail_data.head()
```

```python
#checking for missing values
print(raw_mail_data.isnull().sum())
```

```python
raw_mail_data.describe()
```

```python
raw_mail_data['Category'].value_counts(dropna=False)
```

```python
#label encoding
raw_mail_data.replace({'Category':{'ham':0, 'spam':1}}, inplace=True)
raw_mail_data.head()
```

```python
#separating the data as text and label
X = raw_mail_data['Message']
Y = raw_mail_data['Category']
```

```python
print(X)
```

```python
print(Y)
```

```python
#splitting the data into training data and test data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
X.shape, X_train.shape, X_test.shape
```

```python
#Converting the text data to feature vectors
feature_extraction = TfidfVectorizer(min_df=1, stop_words='english', lowercase=True)
X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)

#Converting Y_train and Y_test values as integers
Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
print(X_test_features)
```

```python
#Training the model: Logistic Regression
model = LogisticRegression()
model.fit(X_train_features, Y_train)
```

```python
#predicting on train data
train_data_prediction = model.predict(X_train_features)
accuracy_train_data = accuracy_score(Y_train, train_data_prediction)
print('Accuracy on training data : ', accuracy_train_data)
```

```python
#predicting the accuracy on the test data
test_data_prediction = model.predict(X_test_features)
accuracy_test_data = accuracy_score(Y_test, test_data_prediction)
print('Accuracy on test data : ', accuracy_test_data)
```

```python
#Building a predictive system
input_mail = ["Congratulations! You've won a free ticket to the Bahamas. Click here to claim your prize now!"]
#convert text to feature vectors
input_mail_features = feature_extraction.transform(input_mail)
#making prediction
prediction = model.predict(input_mail_features)
print(prediction)
```
